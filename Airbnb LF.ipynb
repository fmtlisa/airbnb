- Qu2: Which hosts are the busiest and why?

The busiest hosts will be the hosts with the most listings/ the biggest amount of airbnbs, because they have to take care and host all of those airbnbs. In fact, hosting an airbnb is a lot of work: cleaning, taking care of what may be broken, feys, find clients, fill the descriptions, welcome clients, take care of paiments...

List of the 15 hosts with the more airbnbs listed (by their ID)

```python
besthosts= airbnb[['calculated_host_listings_count','host_id']].sort_values('calculated_host_listings_count', ascending= False).drop_duplicates().head(15)
besthosts
```

Two hosts, 219517861 and 107434423m stand out as having the more having the biggest number of airbnbs listed.

```python
g = sns.catplot(x='host_id', y='calculated_host_listings_count',data=besthosts, saturation=.8,
                kind="bar", aspect=2.5, palette ="ch:s=.25,rot=-.25", order=besthosts.sort_values('calculated_host_listings_count').host_id)

g.set_axis_labels('ID of the 15 hosts with the more listings','Number of listings')
g.fig.suptitle('Amount of listings by the 15 biggest hosts in NYC',y=1.05)
```

This part has been edited in order to do the presentation, based in Joelma's work (not in raw data sent as homework)

```python
airbnb['busy']=(airbnb['calculated_host_listings_count']*airbnb['availability_365'])
besthosts=airbnb.groupby(by=['host_id'])['busy'].sum().sort_values(ascending= False).drop_duplicates().head(15)
besthosts_df=pd.DataFrame(besthosts)
besthosts_df.reset_index(inplace=True)
besthosts_df.rename(columns={'host_id':'host_id', 'busy':'business'}, inplace=True)
besthosts_df
```

```python
g = sns.catplot(x='host_id', y='business',data=besthosts_df, saturation=.8,
                kind="bar", aspect=2.5, palette ="ch:s=.25,rot=-.25", order=besthosts_df.sort_values('business').host_id)

g.set_axis_labels('ID of 15 hosts','Level of business')
g.fig.suptitle('Order of the 15 hosts in airbnb NYC with the most business',y=1.05)
```
